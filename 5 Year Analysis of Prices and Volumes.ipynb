```python
import pandas as pd
```

```python
df = pd.read_csv('Stock Market Dataset.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.shape
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['Bitcoin_Price'] = df['Bitcoin_Price'].str.replace(',', '').astype(float)
df['Platinum_Price'] = df['Platinum_Price'].str.replace(',', '').astype(float)
df['Ethereum_Price'] = df['Ethereum_Price'].str.replace(',', '').astype(float)
df['S&P_500_Price'] = df['S&P_500_Price'].str.replace(',', '').astype(float)
df['Nasdaq_100_Price'] = df['Nasdaq_100_Price'].str.replace(',', '').astype(float)
df['Berkshire_Price'] = df['Berkshire_Price'].str.replace(',', '').astype(float)
df['Gold_Price'] = df['Gold_Price'].str.replace(',', '').astype(float)

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df.set_index('Date', inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['S&P_500_Price'], label='S&P 500 Price')
plt.plot(df.index, df['S&P_500_Price'].rolling(window=5).mean(), label='Weekly (5-Trading Day) Moving Average')
plt.legend()
plt.title('S&P 500 Price and Weekly (5-Trading Day) Moving Average')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df['S&P_500_Price'], model='additive', period=5)
plt.subplot(221)
plt.plot(df.index, df['S&P_500_Price'], label='Original')
plt.legend()

plt.subplot(222)
plt.plot(df.index, decomposition.trend, label='Trend')
plt.legend()

plt.subplot(223)
plt.plot(df.index, decomposition.seasonal, label='Seasonal')
plt.legend()

plt.subplot(224)
plt.plot(df.index, decomposition.resid, label='Residual')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df['S&P_500_Price'], model='additive', period = 20)

plt.figure(figsize=(12, 8))
plt.subplot(221)
plt.plot(df.index, df['S&P_500_Price'], label='Original')
plt.legend()

plt.subplot(222)
plt.plot(df.index, decomposition.trend, label='Trend')
plt.legend()

plt.subplot(223)
plt.plot(df.index, decomposition.seasonal, label='Seasonal')
plt.legend()

plt.subplot(224)
plt.plot(df.index, decomposition.resid, label='Residual')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.api import AutoReg
import pandas as pd
sp_500_price_series = pd.Series(df['S&P_500_Price'].values[::-1])
model = AutoReg(sp_500_price_series, lags=5)
results = model.fit()
print(results.summary())
forecast = results.predict(start=len(sp_500_price_series), end=len(sp_500_price_series) + 9, dynamic=False)
print(forecast)
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(sp_500_price_series, order=(5,1,0))
results = model.fit()
print(results.summary())
forecast = results.forecast(steps=10)
print(forecast)
```
